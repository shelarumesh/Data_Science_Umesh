```python
X = df[['Newspaper','Radio','TV']]
y = df['Sales']
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=16)
```

#Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)

# Coeficient and Intercept
print("Coeficirnts : ", model.coef_)
print("Intercept : ", model.intercept_)

#Prediction
ytrain_pred = model.predict(X_train)
ytest_pred = model.predict(X_test)

#Evolution
lr_train_accuracy = model.score(X_train,ytrain_pred)
lr_test_accuracy = model.score(X_test,ytest_pred)
print("la_train_accuracy : ",lr_train_accuracy)
print("la_test_accuracy : ",lr_test_accuracy)
#Cross validation
cross_validation = cross_val_score(model,X = X_test,y = ytest_pred, cv=5).mean()
print('la_cross_validation : ',cross_validation)
```
