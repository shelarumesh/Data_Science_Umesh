```python
import matplotlib.pyplot as plt
import numpy as np
```

E-01

```python
x=np.linspace(0,10,100)
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
plt.style.use('classic')
fig=plt.figure()
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
%matplotlib inline
plt.style.use('classic')
fig=plt.figure()
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

===

```python
plt.figure()

plt.subplot(2,1,1)
plt.plot(x, np.sin(x))

plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
```

multiple line plot

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.sin(x-1))
plt.plot(x, np.sin(x-2))
plt.plot(x, np.sin(x-3))
plt.plot(x, np.sin(x-4))
plt.plot(x, np.sin(x-5))
```

```python
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':'); # dotted
```
