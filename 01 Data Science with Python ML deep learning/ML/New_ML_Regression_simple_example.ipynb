# New ML Regression simple example

```python
import pandas as pd
df=pd.DataFrame({'x':[1,2,3,4,5,6], 'y':[11,12,13,14,15,16]})
df
```

```python
x=df[['x']]
y=df['y']
```

```python
x,y
```

# Modeling

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

# Prediction

substitute x valye in equation
y =1x+10

```python
ypred = lr.predict(x)
ypred
```

# Evaluation

error, sqare erroe, MSE, RMSE

```python
from sklearn.metrics import mean_squared_error
mse= mean_squared_error(ypred,y)
mse
```

```python
rmse=mse**0.5
rmse
```
