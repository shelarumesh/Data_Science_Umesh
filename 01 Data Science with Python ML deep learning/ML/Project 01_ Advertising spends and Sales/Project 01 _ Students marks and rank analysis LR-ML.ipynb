# 4.Modelling

Train test spit
regression modelling

```python
X=df[['TOTAL SCORE (OUT of 100)']]
y=df['PERCENTILE']
```

```python
from sklearn.model_selection import train_test_split
X_train,x_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()

model.fit(X_train, y_train)
```

```python
print(" coeficient of model is : " ,model.coef_)

print(" Intercept of model is : " ,model.intercept_)
```

# 4.Prediction & Evaluation

```python
y_pred = model.predict(x_test)
```

```python
model.score(x_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mes=mean_squared_error(y_test,y_pred)
print('Mean square erroe of above model is : ', mes)
print('Root mean square erroe of above model is : ', mes**0.5)
```
