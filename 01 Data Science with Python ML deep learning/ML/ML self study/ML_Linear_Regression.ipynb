# 1. Simple Linear Regression

```python
df = pd.DataFrame({'x' :[1,2,3,4,5], 'y':[11,12,13,14,15]})
df
```

```python
X=df[['x']]
y=df.y
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,y)
```

```python
ypred = model.predict(X)
df['ypred'] = ypred
df
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
mae = mean_absolute_error(y,ypred)
mse=mean_squared_error(y,ypred)
rmse = np.sqrt(mse)
print("Mean Absolute erroe is : ", mae)
print("Mean square erroe is : ", mse)
print("Root Mean square erroe is : ", rmse)
model.score(X,y)
```

# regression **line**

```python
plt.scatter(x='x', y='y', data=df)
```

```python
plt.scatter(x='x', y='y', data=df)
plt.scatter(x='x', y='ypred', data=df)
plt.plot(df['x'],df['ypred'])
```

# 2. Simple Non Linear Regression

```python
#non linear dataset
snl = pd.DataFrame({'x' :[1,20,3,4,5], 'y':[9,12,14,14,15]})
snl
```

```python
X=snl[['x']]
y=snl.y
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,y)
```

```python
y_pred = model.predict(X)
snl['y_pred'] = y_pred
snl
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
mae = mean_absolute_error(y,y_pred)
mse=mean_squared_error(y,y_pred)
rmse = np.sqrt(mse)
print("Mean Absolute erroe is of simple non linear regression: ", mae)
print("Mean square erroe is : ", mse)
print("Root Mean square erroe is : ", rmse)
model.score(X,y)
```

```python
plt.scatter(x='x', y='y', data=snl)
plt.scatter(x='x', y='y_pred', data=snl)
plt.plot(snl['x'],snl['y_pred'])
```
