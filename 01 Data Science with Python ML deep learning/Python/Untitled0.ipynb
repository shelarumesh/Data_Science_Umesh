# List Conprehention

method -01

```python
list=[]

def sq_of_all_list(lis):
  for i in lis:
    list.append([i*i])
  return list
```

```python
l=[1,2,3,4,5,6]
sq_of_all_list(l)
```

method 02

```python
[i*i for i in l]
```

```python
[i*i for i in l if i%2==0]

# print only even number from given list
```

```python
[i*i for i in l if i%2!=0]
```
