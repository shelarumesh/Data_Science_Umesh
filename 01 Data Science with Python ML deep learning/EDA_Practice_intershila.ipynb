some data points are see as outlier which is above 6

```python
q1=df['Sale Price'].quantile(0.25)
q3=df['Sale Price'].quantile(0.75)
iqr=q3-q1
iqr
```

```python
lower_limit = q1-(1.5*iqr)
lower_limit
```

```python
upper_limit = q3+(1.5*iqr)
upper_limit
```

```python
l_limit=min(df['Sale Price'])
l_limit
```

lower limit is re define because not matching

```python
plt.hist(df['Sale Price'], bins=10, color='green')
```

```python
#outlier = bet lower and upper limit of data
```

```python
def limit_imputer(value):
  if value > upper_limit:
    return upper_limit
  if value < l_limit:
    return l_limit
  else :
    return value
```

```python
df['Sale Price'].describe()
```

```python
df['Sale Price'].describe()
```

```python
df['Sale Price'] = df['Sale Price'].apply(limit_imputer)
df['Sale Price'].describe()
```
